## DataFrames

Tabular DS
Can be n.m dimensional

```python
import pandas as pd

d1 = pd.DataFrame({"Name": ["Max", "Monica", "Ella", "Matt", "Amy"], "Scores": [99,95,90,87,85]}) #.DataFrame() is a constructor, and not a funcn., hence first letter capital
print(d1)
print(d1.shape)

```

```python
d2 = pd.DataFrame({"Name": ["Max", "Monica", "Ella", "Matt", "Amy"], "Score": [99,95,90,87,85]}, index=["1", "2", "3","4","5"]) #.DataFrame() is a constructor, and not a funcn., hence first letter capital
print(d2)
print(d2.shape)

marksMax = d2.loc["1"]
print(marksMax)

studName = str(input(f"\nEnter your ID: "))
print(f"\nYour score is {d2.at[studName, 'Score']}")
```

### Adding a column

```python
d2["Age"] = [21,22,21,22,20]

print(d2)
```

### Adding a new row

```python
#Create new DFs and concatenate

d2TempRow = pd.DataFrame([{"Name" : "Ron", "Score" : 89, "Age" : 22}, {"Name" : "Mia", "Score" : 88, "Age" : 23}], index = (6,7)) #can add new dicts., for each new row

d2 = pd.concat([d2, d2TempRow])

print(d2)
# print(d2.describe())
print(d2.info)
```

```python
print(d2.info())
print(d2.describe())
```
