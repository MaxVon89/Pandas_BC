##Series

```python
import pandas as pd

dt1 = [1,2,3,4,5]
s1= pd.Series(dt1) #'Series' is a constructor, hence the first letter is capitalized
print(s1)
print(s1.shape)

# dt2 = ['a', 'b', 'c']
dt2 = [['a', 'b', 'c'], ['d', 'e', 'f']]
s2 = pd.Series(dt2)
print(f'\n{s2}')
print(f'{s2.shape}')

print("\nSeries are always 'n' dimensional, not n.m dimensional")
```

```python
s3 = pd.Series([1,2,3])
print(s3)

#directly creating series

s4 = pd.Series([1,2,3], index=['a', 'b', 'c']) #custom indices
print(f"\n{s4}")
```

```python
#practical usage

seriesAddress = pd.Series(['RiverHouse', '87', 'Lane 9', 'Aliston Rd', 'Palm Springs', 'CA'], index=['Name (optional):','Plot no.:', 'Lane No.:', 'Street:', 'City:', 'State\\Province:'])
print(seriesAddress)

```

---
loc & iloc

loc searches for indices, by their 'given' name;
whereas,
iloc searches them by their actual index.

```python
#loc

seriesAddress.loc["Name (optional):"] = "Anatolia"
print(seriesAddress)

print(f"\n{seriesAddress.loc["City:"]}")
```

```python
# iloc

seriesAddress.iloc[1] = 95
print(seriesAddress)

print(f"\n{seriesAddress.iloc[5]}")
```

```python
studentsAndScores = pd.Series([99,97,88,85,87], index=['Max', 'Monica', 'Ella', 'Matt', 'Amy'])
print(studentsAndScores)

excellentStudents = studentsAndScores[studentsAndScores >= 95]
print(f"\n{excellentStudents}")

```

```python
#Dicts. as Series

studentsAndScores2 = pd.Series({'Max':99, 'Monica':98, 'Ella':85, 'Matt':89, 'Amy':88})
print(studentsAndScores2)
#no need to define indices separately, this way.

excellentStudents2 = studentsAndScores2[studentsAndScores2 >= 95]
print(f"\n{excellentStudents2}")

#loc here:
MaxScores = studentsAndScores2.loc['Max']
MonicaScores = studentsAndScores2.loc['Monica']
EllaScores = studentsAndScores2.loc['Ella']
print(f"\n{MaxScores, MonicaScores, EllaScores}")


MaxScores2 = studentsAndScores2.iloc[0]
MonicaScores2 = studentsAndScores2.iloc[1]
EllaScores2 = studentsAndScores2.iloc[2]
print(f"\n{MaxScores, MonicaScores, EllaScores}")


#Let's bump up Ella's score by 5 marks, and print it.

studentsAndScores2.loc["Ella"] =+5
print(f"\nElla's new score: {studentsAndScores2.loc['Ella']}")
```
