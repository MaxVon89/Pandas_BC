## Aggregate Functions

Reduce a set of vals in a single summary val.
Used to summarize, organize data
Often used w/the groupby() function

```python

import pandas as pd

d10 = pd.read_csv("D:\\Datasets\\multipleLinearRegression\\CarPrice_Assignment.csv")

d10
```

### Whole DataFrame Aggr. Funcs

```python
# d10.mean()


wholeMean, wholeSum, wholeMin, wholeMax, wholeCount = d10.mean(numeric_only=True), d10.sum(numeric_only=True), d10.min(numeric_only=True), d10.max(numeric_only=True), d10.count()
# wholeMean, wholeSum, wholeMin, wholeMax, wholeCount = d10.mean(numeric_only=True), d10.sum(numeric_only=True), d10.min(numeric_only=True), d10.max(numeric_only=True), d10.count(numeric_only=True)
print(f"{wholeMean}\n\n{wholeSum}\n\n{wholeMin}\n\n{wholeMax}\n\n{wholeCount}")
```

### Single Column Aggr. Funcs.

```python
colMean = d10['carlength'].mean()
colSum = d10['carlength'].sum()
colMin = d10['carlength'].min()
colMax = d10['carlength'].max()
colCount = d10['carlength'].count()

print(f"\n{colMean, colSum, colMin, colMax, colCount}")
colMean, colSum, colMin, colMax, colCount
```

## GroupBy()

```python
# groupItem = input("Enter group atrr.: ")
groupItem = 'carbody' #default
aGroup = d10.groupby(groupItem)
# aGroup.head() #Not useful here

print(aGroup.sum(numeric_only=True))
# print(aGroup.count(numeric_only=True))
```
